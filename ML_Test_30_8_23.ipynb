Binning


```python
bins = [0,4,5.5,8]
bin_labels=['Low Happiness','Medium Happiness','High Happiness']
happiness_count=pd.cut(data['Happiness Score'],bins=bins,right=False).value_counts().sort_index()
plt.bar(bin_labels,happiness_count,color=['green','blue','red'])
plt.show()
```

One-Hot Encoding

```python
pd.unique(data['Region'])
#identifying how many unique values are there in the region column
```

```python
data_encoded=pd.get_dummies(data,columns=['Region'])
#dummy encoding is producing too many unusable columns,
#so we can only use one hot encoding
```

```python
#Creating a contingency table using contingency table
from scipy.stats import chi2_contingency

contingency_table=pd.crosstab(data['Region'],data['Happiness Rank'])
chi2,p,dof,frequency=chi2_contingency(contingency_table)
print("Chi-Square Test Results : ")
print('Chi-Sqaure Value : ',chi2)
print('P-Value : ',p)
print('Degrees of freedom: ',dof)
print('Frequency Distribution : ',frequency)

#as seen below, the value of p is below 0.05 and null hypothesis is rejected
```

```python
covariance_matrix = np.cov(data[['Economy (GDP per Capita)','Health (Life Expectancy)']],rowvar=False)
covariance_matrix
```
